```python
import requests, os, time, csv, re, os, time
from bs4 import BeautifulSoup
```

```python
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.75 Safari/537.36'}
```

```python
def parse_haodf(html,page):

    filtered = []
    one = html
    two = '.htm'
    url_list = []
    for i in range(1,page+1):
        url_curr = one + str(i) + two
        url_list.append(url_curr)
    
    for i in url_list:
        time.sleep(2)
        r = requests.get(i, headers=headers).text
        soup = BeautifulSoup(r, 'lxml', from_encoding='utf-8')
        result = soup.find_all('div')
        for j in result:
            try:
                if j.get('class')[0] ==  'advise_box_title':
                    if '.html'in j.find('a').get('href'):
                        filtered.append(j.find('a').get('href'))
            except:
                print(j)
                
    return filtered
```

```python
def parse_chunyu(html,page):

    filtered = []
    one = html
    url_list = []
    for i in range(1,page+1):
        url_curr = one + str(i)
        url_list.append(url_curr)
    
    
    for i in url_list:
        time.sleep(2)
        r = requests.get(i, headers=headers).text
        soup = BeautifulSoup(r, 'lxml', from_encoding='utf-8')
        result = soup.find_all('div')
        for j in result:
            try:
                if j.get('class')[0] ==  'qa-item' and j.get('class')[1] == 'qa-item-ask':
                    if '/pc/qa'in j.find('a').get('href'):
                        filtered.append('https://www.chunyuyisheng.com' + j.find('a').get('href'))
                        
            except:
                print(j)
        
        time.sleep(2)
      
                
    return filtered
```

```python
def parse_39(html,page):

    filtered = []
    one = html
    two = '.htm'
    url_list = []
    for i in range(1,page+1):
        url_curr = one + str(i)
        url_list.append(url_curr)
    
    for i in url_list:
        time.sleep(2)
        r = requests.get(i, headers=headers).text
        soup = BeautifulSoup(r, 'lxml', from_encoding='utf-8')
        result = soup.find_all('a')
        for j in result:
            
            try:
                
                reg_date = r'"http://ask.39.net/question/([\s\S]+?)"'
                pattern_date = re.compile(reg_date)
                found = re.findall(pattern_date,r)
                for k in found:
                    to_add = 'http://ask.39.net/question/' + k
                    if to_add not in filtered:
                        filtered.append(to_add)
            except:
                pass
                
    return filtered
```

```python
def parse_120(html,page):

    filtered = []
    one = html
    two = '&isloc=1'
    url_list = []
    for i in range(1,page+1):
        url_curr = one + str(i)
        url_list.append(url_curr)
    
    for i in url_list:
        time.sleep(2)
        r = requests.get(i, headers=headers).text
        soup = BeautifulSoup(r, 'lxml', from_encoding='utf-8')
        result = soup.find_all('a')
        for j in result:
            
            try:
                
                reg_date = r'"https://www.120ask.com/question/([\s\S]+?)"'
                pattern_date = re.compile(reg_date)
                found = re.findall(pattern_date,r)
                for k in found:
                    to_add = 'https://www.120ask.com/question/' + k
                    if to_add not in filtered:
                        filtered.append(to_add)
            except:
                pass
                
    return filtered
```

```python
def analyze_haodf(url_list,disease):
    unchecked = []
    for i in range(len(url_list)):
    
        time.sleep(2)
        try:
            full_text = []
            url = url_list[i]
            forma = [ '提示：任何关于疾病的建议','好大夫在线提供','微信扫码关注医生','有问题随时问', '综合推荐热度','在线服务满意度','在线问诊量','推荐热度','大夫的信息','简介','擅长','相关文章','相关回复','相关疾病']
            r = requests.get(url, headers=headers).text
            soup = BeautifulSoup(r, 'html5lib', from_encoding='utf-8')
            result = soup.find_all('p')
            for j in result:
                to_add = ''
                for char in j.getText():
                    if char != ' ':
                        to_add = to_add + char
                allow = 1
                for element in forma:
                    if element in to_add:
                        allow = 0
                if allow == 1:
                    to_add = to_add.replace('\n','')
                    full_text.append(to_add)
            full_text = ''.join(full_text)
                
        
            if not os.path.exists(disease):
                os.makedirs(disease)
            address = disease + '\\' + disease + '-haodf-' + str(i) + '.txt'
            with open(address,'w',encoding='utf-8') as text:
                text.write(full_text)
            
        except:
            unchecked.append(url_list[i])
        
        
        if i%5 == 0:
            print(str(round((100*i/len(url_list)),2)) + '%')
        
    return unchecked
        
```

```python
def analyze_39(url_list,disease):
    unchecked = []
    for i in range(len(url_list)):
    
        time.sleep(2)
        try:
            url = url_list[i]
            r = requests.get(url, headers=headers).text
            soup = BeautifulSoup(r, 'html5lib', from_encoding='utf-8')
            result = soup.find_all('p')
    
            full_text = []
            for j in result:
                include = ['txt_ms', 'ask_tit','sele_txt']
                if j.get('class') != None:
                    if j.get('class')[0] in include:
                        to_add = j.getText()
                        full_text.append(to_add)
            full_text = ''.join(full_text)

            
            
            if not os.path.exists(disease):
                os.makedirs(disease)
            address = disease + '\\' + disease + '-39-' + str(i) + '.txt'
            
            with open(address,'w',encoding='utf-8') as text:
                text.write(full_text)
          
        except:
            unchecked.append(url_list[i])
        
        
        if i%5 == 0:
            print(str(round((100*i/len(url_list)),2)) + '%')
        
    return unchecked
```
